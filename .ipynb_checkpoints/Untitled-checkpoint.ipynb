```python
import numpy as np #for generating random

def gen_ran(n,f):
    #generate an array of n random numbers
    a = []
    for i in range(n):
        a.append(np.random.exponential(f))
    return a
```

# Test run

```python
to_come = 20 #number of customers to come
arrival_size = 1
service_size = 1
servers = 2 #number of servers
time_unit =1/100 #the minimal time interval in simulation
```

```python
#get the interval of arrival time
arrival_interval = gen_ran(to_come,arrival_size)

#transform arrival interval into arrival time
arrival_time = []
t = 0
for i in arrival_interval:
    t += i
    arrival_time.append(t)

#get the interval of service time
service_interval = gen_ran(to_come,service_size)

print(arrival_time) 
print(service_interval)
```

```python
#servers
class Servers(object):
    def __init__(self,initial, time_unit):
            self.serving = np.array(initial)
            self.time_unit = time_unit
            self.refresh()
    
    def refresh(self):
        #refresh the self.vacant
            self.vacant = np.array([t <= 0 for t in self.serving])
    
    def find_vacancies(self):
        #return the id of the first vacancy
            return np.array([(t == 0) for t in self.serving])
        
    def find_to_finish(self):
        #return an array of whether server is to finish in the next round
            return np.array([(t <= self.time_unit) and (t>0) for t in self.serving])
            
    def update(self, server_id, service_time):
        #change the service time of a specified server
            self.serving[server_id] = service_time
            self.refresh()

    def count_down(self):
        #to count down the time by one unit
            self.serving = np.array([max(0, t - self.time_unit) for t in self.serving])
            self.refresh()

s = Servers([1,1,2,1,0], 0.5)
print(s.serving)
s.count_down()
print(s.serving, s.find_vacancies(),s.find_to_finish())
s.count_down()
print(s.serving, s.find_vacancies(),s.find_to_finish())
s.count_down()
print(s.serving, s.find_vacancies(),s.find_to_finish())
```
